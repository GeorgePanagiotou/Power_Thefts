```python
for i in range(len(train)) :
  train.at[i, 'START_DATE'] = train.at[i, 'START_DATE'].replace("-","")
#train_data.head()
```

```python
est = KBinsDiscretizer(n_bins=1000, encode='ordinal', strategy='uniform')
train['START_DATE'] = est.fit_transform(np.array(train['START_DATE']).reshape(-1,1))
#train_data.head()
```
